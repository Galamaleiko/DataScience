```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

dataSet = pd.read_excel("rollingsales_statenisland.xls", skiprows=4)
dataSet
```

```python
dataSet.info()
```

```python
dataSet.isnull().sum()
```

```python
dataSet["YEAR BUILT"][dataSet["YEAR BUILT"] == 0] = np.NaN
dataSet["SALE\nPRICE"][dataSet["SALE\nPRICE"] == 0] = np.NaN
dataSet["GROSS SQUARE FEET"][dataSet["GROSS SQUARE FEET"] == 0] = np.NaN

sb.heatmap(dataSet.isnull(),cbar=False,yticklabels=False)
print("Поглянемо на heatMap, чи залишились у нас нульові або null-ові значення")
```

```python
dataSet.dropna(axis=0,inplace=True)

print("Оскільки було виявлено непідходящі для аналітики значення, ми змушені їх прибрати.")
sb.heatmap(dataSet.isnull(),cbar=False,yticklabels=False)
```

```python
sb.set(rc={"figure.figsize": (7,9)})
sb.countplot(data=dataSet,y="NEIGHBORHOOD", order= dataSet["NEIGHBORHOOD"].value_counts().index).set(title="sales on district")
print("Оглядаємо ціни по районах")
```

```python
order = dataSet.groupby(["NEIGHBORHOOD"])["RESIDENTIAL UNITS"].mean().sort_values().index
sb.barplot(y="NEIGHBORHOOD", x="RESIDENTIAL UNITS",data=dataSet, order=order).set(title="Residental sales")
print("Оглядаємо ціни по районах на житлові будівлі")
```

```python
order = dataSet.groupby(["NEIGHBORHOOD"])["COMMERCIAL UNITS"].mean().sort_values().index
sb.barplot(y='NEIGHBORHOOD',x='COMMERCIAL UNITS',data=dataSet,errorbar=None, order=order).set(title="Commercial Sales")
print("Оглядаємо ціни на комерційні будівлі")
```

```python
order = dataSet.groupby(["NEIGHBORHOOD"])["YEAR BUILT"].mean().sort_values().index
sb.barplot(y='NEIGHBORHOOD',x='YEAR BUILT',data=dataSet,errorbar=None, order=order).set(title="Group by Year Built")
print("Бачимо, що старих забудівель немає")
```

```python
building_class_category=dataSet.groupby("BUILDING CLASS CATEGORY")[["SALE\nPRICE"]].mean()
building_class_category.plot(kind='bar',xlabel="Class Category",ylabel="Avg price")
print("Оцінюємо середню ціну, розподілену по категоріях")
```
